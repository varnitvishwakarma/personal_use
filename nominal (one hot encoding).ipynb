```python
# one hot encoding
```

```python
import pandas as pd
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
df=pd.DataFrame({"color":["blue","black","black","blue","green","green"]})
```

```python
# create an instance of onehotencoder
```

```python
encoder=OneHotEncoder()
```

```python
encoded=encoder.fit_transform(df[["color"]])
```

```python
encoded_data=pd.DataFrame(encoded.toarray(),columns=encoder.get_feature_names_out())
```

```python
encoder.get_feature_names_out()
```

```python
pd.concat([df,encoded_data],axis=1)
```

```python
# one hot encoding on tips data set
```

```python
import seaborn as sns
```

```python
tips_data=sns.load_dataset("tips")
```

```python
tips_data.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
Encoder=OneHotEncoder()
```

```python
encoded_tips_data=Encoder.fit_transform(tips_data[["sex","smoker","day","time"]])
```

```python
pd.DataFrame(encoded_tips_data.toarray(),columns=Encoder.get_feature_names_out())
```
